## Проверка гипотез
[Ссылка к разделу план работы](#section_id_10)

**Проверим гипотезу "средние пользовательские рейтинги платформ Xbox One и PC одинаковые"**

Так как нам необходимо проверить средний рейтинг между двумя платформами, а именно равны ли они, то мы будем опираться опровергнем мы нулевую гипотезу или нет, а альтернативную гипотезу поставим ей в противовес как не равны.

H_0: средние пользовательские рейтинги `user_score` платформы Xbox One = средние пользовательские рейтинги `user_score` платформы PC

H_1: средние пользовательские рейтинги `user_score` платформы Xbox One != средние пользовательские рейтинги `user_score` платформы PC

Так как это независимые выборки то применим двухвыборочный t-тест для независымых выборок (ttest.ind)

```python
# сделаем две выборки для каждой из платформ и избавимся от нулевых значений, чтобы они не искажали среднее 
sample_xone = df_actual[(df_actual['platform']=='xone') & (df_actual['user_score']!=-1)]
sample_pc = df_actual[(df_actual['platform']=='pc') & (df_actual['user_score']!=-1)]
# проверим дисперсию
print(np.var(sample_xone['user_score']), np.var(sample_pc['user_score']))
```

Отношение большей дисперсии выборки к меньшей дисперсии выборки не различаются в 4 раза, мы можем предположить, что дисперсии генеральной совокупности равны.

Таким образом, мы можем приступить к выполнению двухвыборочного t-критерия с равными дисперсиями

```python
# выбранный уровень статистической значимости
alpha = 0.05

results_1 = st.ttest_ind(sample_xone['user_score'], sample_pc['user_score'], equal_var = True)

# выведем среднее время сессии для пользователей обоих типов подписок
print('Средняя оценка пользователей на платформе xone:', round(sample_xone['user_score'].mean(), 2))
print('Средняя оценка пользователей на платформе pc:', round(sample_pc['user_score'].mean(), 2))
print()
# выведем полученное p-value
print('p-значение', results_1.pvalue)
print()
# вывод об отвержении или неотвержении нулевой гипотезе
if results_1.pvalue < alpha:
    print('Отвергаем нулевую гипотизу')
else:
    print('Не отвергаем нулевую гипотизу')
```

Гипотезу "средние пользовательские рейтинги платформ `Xbox One` и `PC` одинаковые" при уровне статистической значимости: 5% нельзя опровергнуть они достаточно близки.

**Проверим гипотезу "средние пользовательские рейтинги жанров `Action` и `Sports` разные"**

Так как нам необходимо проверить средний рейтинг между двумя платформами, а именно то, что они разные, то опираемся мы на  альтернативную гипотезу которую мы поставим, что они не равны, а нулевая гипотеза у нас выставляется, что они одинаковые.

H_0: средние пользовательские рейтинги `user_score` жанра `Action` = средние пользовательские рейтинги `user_score` жанра `Sports`

H_1: средние пользовательские рейтинги `user_score` жанра `Action` != средние пользовательские рейтинги `user_score` жанра `Sports`

Так как это независимые выборки то применим двухвыборочный t-тест для независымых выборок (ttest.ind)

```python
# сделаем две выборки для каждого из жанров и избавимся от нулевых значений, чтобы они не искажали среднее 
sample_action = df_actual[(df_actual['genre']=='action') & (df_actual['user_score']!=-1)]
sample_sports = df_actual[(df_actual['genre']=='sports') & (df_actual['user_score']!=-1)]
# проверим дисперсию
print(np.var(sample_action['user_score']), np.var(sample_sports['user_score']))
```

Отношение большей дисперсии выборки к меньшей дисперсии выборки не различаются в 4 раза, мы можем предположить, что дисперсии генеральной совокупности равны.

Таким образом, мы можем приступить к выполнению двухвыборочного t-критерия с равными дисперсиями

```python
# выбранный уровень статистической значимости
alpha = 0.05

results_2 = st.ttest_ind(sample_action['user_score'], sample_sports['user_score'], equal_var = True)

# выведем среднее время сессии для пользователей обоих типов подписок
print('Средняя оценка пользователей в жанре Action:', round(sample_action['user_score'].mean(), 2))
print('Средняя оценка пользователей в жанре Sports:', round(sample_sports['user_score'].mean(), 2))
print()
# выведем полученное p-value
print('p-значение', results_2.pvalue)
print()
# вывод об отвержении или неотвержении нулевой гипотезе
if results_2.pvalue < alpha:
    print('Отвергаем нулевую гипотизу')
else:
    print('Не отвергаем нулевую гипотизу')
```

Гипотезу "средние пользовательские рейтинги жанров `Action` и `Sports` разные" при уровне статистической значимости: 5% нельзя опровергнуть, средние оценки достаточно разные.

**Вывод**

Если обратиться к источника в интернете то выясниться, что ESRB - негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде и логично, что для других регионов он не заполняется, также она была основана в 1994 году.

Можно предположить, что пропуски связаны с тем в Японии есть свой рейтинг, свой рейтинг есть в ЕС и отдельно в Германии, Австралии и т.д., но по ним в датасете нет данных, а также что игры до 1994 года, не могли получить этот рейтинг, также игры выпускаются не толко в регионе где присутствует данная организация. Игры могли быть малоизвестные, не популярность жанра или платформы, игры были выпущенны для определенного регионна, ну или просто не попали в поле зрения данной организации.

**Резюмируя:**

Если успех игры исчислять количеством проданных копий, то основные параметры влияющие на объём продаж это:

* жанр;
* платформа;
* возрастной рейтинг;
* регион.

Суммарно, по эти 4-ём параметрам, можно подсчитать ожидаемый диапазон продаж.

Например:

`Shooter` или `Action` с возрастным рейтингом `M` или `E` для домашних консолей `Playstation 4` - прибыльный вариант для территории Европы. Аналогичный вариант подойдёт Северной Америке, с дополнительным каналом рекламы: Xbox One. Эта приставка одинакова популярна как PS4.
`Role-Playing` или `Action` исключительно для портативной консоли `Nintendo 3DS`, возрастное ограничение ESRB не имеет значение - верный вариант для Японии.
При выборе можно учесть отзыв критиков. Желательно, чтобы оценка была от 70 баллов. Рейтинг пользователей - нельзя расценивать однозначно.

В перечень параметров игр можно добавить:

* дату релиза;
* стоимость игры (на момент выпуска) (это позволит определить успех игр с финансовой стороны).
